```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection
matplotlib.style.use('seaborn-talk')
%matplotlib inline
```

```python

# Figure and arena
def plot_sham_diagram(ax, obj_radius=5, buffer_zone=5, sham_shape=(50, 10), wallhug_dist=10, n_shams=8, sham_alpha=.5):

    rect = plt.Rectangle((-50, -25), 100, 50, fill=True, linestyle='dotted', linewidth=3, color='y', alpha=.3)
    ax.add_patch(rect)

    wall_hugging_rect = plt.Rectangle((-50 + wallhug_dist, -25 + wallhug_dist), 
                                      100 - wallhug_dist * 2, 50 - wallhug_dist * 2, 
                                      fill=True, linestyle='dotted', color='w')
    ax.add_patch(wall_hugging_rect)

    # Sham object
    def gen_random_circles(n=8, alpha=.1):
        circles = []
        for el in range(n):
            pos = np.zeros(2)
            while np.linalg.norm(pos) < obj_radius + buffer_zone or np.linalg.norm(pos - [30, 5]) < obj_radius + buffer_zone:
                pos = (np.random.random(2) * 2 - 1) * np.divide(sham_shape, 2)
            circle = plt.Circle(pos, radius=obj_radius, alpha=alpha)
            circles.append(circle)
        return circles
    rect = plt.Rectangle(-np.divide(sham_shape, 2), sham_shape[0], sham_shape[1], fill=False, linestyle='dotted')
    ax.add_patch(rect)

    for circle in gen_random_circles(n_shams, alpha=sham_alpha):
        ax.add_patch(circle)

    # Fixed object
    obj_circle = plt.Circle((0, 0), radius=obj_radius, 
                            color='r', alpha=.6)
    ax.add_patch(obj_circle)

    nogo_circle = plt.Circle((0, 0), radius=obj_radius + buffer_zone, 
                             fill=False, color='r', linestyle='dotted', linewidth=1, alpha=1)
    ax.add_patch(nogo_circle)

    # Second object
    other_circle = plt.Circle((30, 5), radius=obj_radius, color='g', alpha=.6)
    ax.add_patch(other_circle)


    other_nogo_circle = plt.Circle((30, 5), radius=obj_radius + buffer_zone, 
                                   fill=False, color='g', linestyle='dotted', linewidth=1, alpha=1)
    ax.add_patch(other_nogo_circle)

    ax.set(
        xlim=(-60, 60),
        ylim=(-35, 35),
        aspect='equal'
    )
    return ax
```

## Original Randomization Schema and Parameters

```python
fig, ax1 = plt.subplots()#, figsize=(5, 8))
plot_sham_diagram(ax1, obj_radius=5, buffer_zone=5, sham_shape=(50, 10), 
                  wallhug_dist=10, n_shams=8, sham_alpha=.5)
```
