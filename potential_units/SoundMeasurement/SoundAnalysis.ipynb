```python
%matplotlib inline

import os
import mpld3
from IPython.display import display
import numpy as np
import matplotlib.pyplot as plt

from scipy.io import wavfile
from scipy.signal import periodogram

#mpld3.enable_notebook()
plt.ion()
```

#### Calculate Periodogram from Sound Files recorded with Ultramic250k

```python
freq_min, freq_max = 0, 10000
n_fft = 2048 * 128

file_list = [name for name in os.listdir(os.getcwd()) if '.wav' in name]
ref_file = filter( lambda x: 'Fan0_' in x, file_list)[0]
ref_file_idx = file_list.index(ref_file)
display(list(enumerate(file_list)))

print('Reference Sound File is at position {0}: {1}'.format(ref_file_idx, ref_file))

power_list = []
for file_name in file_list:
    
    # Read Data
    samp_rate, data = wavfile.read(file_name)
    data = data.astype(float) / 2**16  # Rescale to percentage of max mic amplitude.
    
    # Calculate Periodogram
    freq, power = periodogram(data, samp_rate, nfft=n_fft)

    # Frequency Cutoff
    power, freq = power[freq < freq_max], freq[freq < freq_max]
    
    power_list.append(power)
    
# Plot an example periodogram (in this case, the last one.)
idx = 6
plt.plot(freq, power_list[idx])
plt.title(file_list[idx])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitute (% of Max Amplitude)')
```

```python
def total_energy(power):
    """Gets some single number out of an array of power values across freqs."""
    tot = np.sum(power)
    #tot = np.log10(tot)
    return tot
    
ref_energy = total_energy(power_list[ref_file_idx])
tot_energy, rel_energy = [], []
for power in power_list:
    temp = total_energy(power)
    
    tot_energy.append(temp) 
    rel_energy.append( temp / ref_energy)
    
# Plot Total Energy in Each File    
plt.figure()    
plt.plot(tot_energy)
plt.xlabel('File Number')
plt.ylabel('Total Amplitude, Across Freqs')

# Plot Relative Energy to Reference File for Each File
plt.figure()
plt.plot(rel_energy)
plt.xlabel('File Number')
plt.ylabel('Relative Amplitude (x1 / x0)')

    
        

```

### Look at Energy increase with Fan Level

```python
fan_levels = 0, 2, 3, 4, 5
energy_levels = []
fan_indices = []
for level in fan_levels:
    name = filter(lambda x: str(level) in x, file_list)
    assert len(name) == 1
    name = name[0]
    idx = file_list.index(name)
    fan_indices.append(idx)
    energy_levels.append(tot_energy[idx])
    
plt.plot(fan_levels, energy_levels, '--o')
plt.xlabel('Fan Speed Setting')
plt.ylabel('Total Energy')
    
```

### Look at Energy Increase in Other Conditions

```python
other_names = []
other_energy = []

# Start with Reference File
other_names.append(ref_file[:-14])
other_energy.append(tot_energy[ref_file_idx])

# Get values for all other files not used in fan speed comparison (above)
for idx, energy in enumerate(tot_energy):
    if idx not in fan_indices:  # Only take files not looking at fan.
        other_names.append(file_list[idx][:-14])
        other_energy.append(energy)
        
# Plot Them
x = np.arange(len(other_energy))

plt.plot(x, other_energy)
plt.xticks(x, other_names, rotation=90) 
plt.ylabel('Total Energy')
       
        
    
```
