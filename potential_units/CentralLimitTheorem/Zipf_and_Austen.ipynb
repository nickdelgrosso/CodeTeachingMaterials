# Zipf's Law

Zip's law is about the relative frequency of occurances: for example, occurances of a word, burst lengths of a neuron, net financial worth in individuals.  It is the observation that if you divide the frequencey of the most-occuring thing from the second-most-occuring thing, you get the same ration as if you divde the second from the third, and so on.  It has to do with Power laws.

## Load the Data

Here it's done for you using the [requests](https://requests.readthedocs.io/en/master/) package. The book's text is downloaded and put into a string called "book".

```python
import requests
url = "https://raw.githubusercontent.com/nickdelgrosso/Intro-to-Python-Course/master/Day1_Code_Blocks/Jane%20Austen___Pride%20and%20Prejudice.txt"
r = requests.get(url)
r.ok
```

```python
book = r.text
type(book)
```

```python
print(book[:400])
```

```python
print(book[-200:])
```

## Seperate all the words from each other
Essentially, turn this book into a list of words

```python
book[:200]
```

#### Get rid of newline characters
### Seperate into words

```python
words = book.split()
print(words[:50])
```

## Count the number of times each word appears in the text
Don't worry about being too exact here (for example, we can treat "shoe" and "shoes" as different words if we want for this exercise), but do make sure you don't count whitespace as words (spaces, newline characters, etc) 

```python
counts = Counter(words)
counts
```

## Make the Plot!

```python
import matplotlib.pyplot as plt
plt.plot(counts.most_common(50))
```
