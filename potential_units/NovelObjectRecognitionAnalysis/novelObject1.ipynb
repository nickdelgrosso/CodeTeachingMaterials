```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

## Load Data into a DataFrame for easy analysis

```python
file_name = 'sampData.csv'
data = pd.io.parsers.read_csv(file_name, sep=',')
print(list(data.columns.values))
```

## Make Histograms of Time exploring each object

```python
fig = plt.figure()
plt.title('Exploration Times of Each Object (seconds)')
columns = ['Novel (s)', 'Sample (s)']
for column in columns:
    data[column].hist(bins=7)
plt.legend(columns)



fig2 = plt.figure()
plt.title('Difference Score (novel - sample)')
(data[columns[0]] - data[columns[1]]).hist(bins=7)
plt.plot([0, 0], [0, 9], 'k')

fig2 = plt.figure()
plt.title('Discrimination Score sample / (novel - sample)')
ratio = (data['Novel (s)'] / (data['Novel (s)'] + data['Sample (s)']))
ratio.hist(bins=7)
plt.plot([0.5, 0.5], [0, 6], 'k')

plt.show()
```
