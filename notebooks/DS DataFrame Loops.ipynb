```python
df2 = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/exercise.csv', index_col=0)
os.makedirs('exercise_data', exist_ok=True)
for (kind, diet, id), group in df2.groupby(['kind', 'diet', 'id']):
    dd = group.drop(labels=['kind', 'diet', 'id'], axis='columns')
    dd.to_csv(f'exercise_data/exercise_{id}_{kind}_{diet}.ignore.csv', index=False)
```

```python
search = 'exercise_data/exercise_{id}_{kind}_{diet}.ignore.csv'
ids, kinds, diets = glob_wildcards(search)
files = [search.format(id=id, kind=kind, diet=diet) for id, kind, diet in zip(ids, kinds, diets)]
df = pd.DataFrame({'id': ids, 'kind': kinds, 'diet': diets, 'file': files })
df
```
