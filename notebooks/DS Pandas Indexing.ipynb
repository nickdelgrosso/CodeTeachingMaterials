## Manging Indices

| Function | Description | Example |
| :-- | :-- | :-- |
| `pd.read_csv(url, index_col=0)` | Read a csv file, setting the index column to a given column number or name | `pd.read_csv(url, index_col='author'`|
| `df.index` | Get the index data |  |
| `df.set_index(col)`| Move a column into the index | `df.set_index('Country')` |
|`df.reset_index()` | Move the index into a new column |  |
| `df.reset_index(drop=True)` | Reset the index without making a new column |  |
| `df.sort_index()` | Sort the rows by the index values |
| `df.loc[name]` | Index the row by its name | `df.loc['Canada']` |
| `df.loc[start:stop]` | slice the rows by index name | `df.loc['Canada':'Germany']` |
| `df.iloc[idx]` | Index the row by its position (like in lists) | `df.iloc[15]` |
| `df.iloc[start:stop]` | Slice the rows by position (like in lists) | `df.iloc[15:20]` |
| **`df.iterrows()`** | iterate over the rows of data  | **`for idx, row in df.iterrows():`** |
